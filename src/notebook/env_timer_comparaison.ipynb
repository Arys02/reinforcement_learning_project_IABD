```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'env': ['lineworld', 'gridworld', 'montyhall', 'tworoundrps', 'secretenv0', 'secretenv1', 'secretenv2', 'secretenv3'],
    'monte_carlo_off': [29, 32, 19, 21, 556, 2257, 2379, 2364],
    'monte_carlo_on': [26, 77, 25, 24, 713, 3274, 3398, 2536],
    'monte_carlo_es': [34, 83, 10, 26, 274, 1219, 2731, 1172],
    'q_learning': [36, 97, 11, 27, 274, 1285, 2936, 1208],
    'sarsa': [45, 1030, 20, 18, 1040, 4375, 8471, 4722]
}
```

```python
df = pd.DataFrame(data)
```

```python
fix, ax = plt.subplots(figsize=(12,8))

# Tracer chaque ligne
for index, row in df.iterrows():
    ax.plot(df.columns[1:], row[1:], label=row['env'])

ax.set_xlabel('Algorithm')
ax.set_ylabel('time millisec')
ax.set_title('Comparison of different environments across algorithms')
ax.legend()

# Afficher le graphique
plt.show()
```
